```python
from time import sleep
from tqdm import tqdm
```

```python
for _ in range(1000):
    sleep(0.001)
```

```python
for _ in tqdm(range(1000)):
    sleep(0.001)
```

```python
from tqdm import trange
for _ in trange(1000):
    sleep(0.001)
```

```python
# 惰性可迭代对象
def my_generator():
    for i in range(50):
        yield i

for _ in tqdm(my_generator(),desc='这是tqdm修饰器的desc参数作用'):
    pass
```

```python
# 惰性可迭代对象
def my_generator():
    for i in range(50):
        yield i

for _ in tqdm(my_generator(),desc='这是tqdm修饰器的desc参数作用',total=50):
    pass
```

```python
# 手动更新tqdm的进度条
with tqdm(total=100) as pbar:
    sleep(10)
    pbar.update(10)
    sleep(10)
    pbar.update(20)
    sleep(10)
    pbar.update(70)
```
